### Numpy en Fourier
In het Fourier domein is het belangrijk om de juiste frequenties aan de data te koppelen. Als voorbeeld onderstaande code waar de [numpy rfft](https://docs.scipy.org/doc/numpy/reference/generated/numpy.fft.rfft.html) routine voor reele input gebruikt wordt:

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

samplerate = 44.1e3 #44.1 kHz

t = np.linspace(0, 4, int(samplerate*4)) # 4 seconden data
data = 3*np.sin(2*np.pi*500*t) # toon van 500 Hz met amplitude 3
data += np.sin(2*np.pi*800*t)
fdata = np.abs(np.fft.rfft(data))*2/(len(data)) #normalisatie van Fourer transfrom
                                # de *2 alleen omdat het een rfft functie is. Het complexe deel is 
                                # numeriek overgeslagen, maar nog steeds nodig voor normalisatie.
freqs = np.fft.rfftfreq(len(data), d=1/samplerate)

fig, (ax1, ax2)  = plt.subplots(2, 1)
ax1.plot(t, data)
ax2.plot(freqs,fdata.real)
plt.show()
````
